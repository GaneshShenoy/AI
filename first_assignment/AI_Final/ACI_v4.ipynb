### Approach to solve given problem

Given problem is solved in two steps. 
1. Selecting the package to load to robot for each commute
2. Selected load from the robot is places in available free room.

For step 1, the cost function is going to be the number of trips robot will take. And based on the input We have total 131kg and each time robot can pick max 10kg so, min cost will be 131/10 i.e. 13 Trips + 1 Trip for remainder of 1 kg so, 14 Trips

For our usecase problem 2 the cost function is going to be the number of rooms used. And based on the input We have total 131kg and total rooms capacity 159kg, space left should be 159-131 i.e. 28kg So, Room 20kg and Room 5kg can be empty, only 4 rooms should be used.  

```python
"""First Fit Decreasing (FFD): First order the items by size, from largest to smallest, then run
the First Fit Algorithm."""

def firstFit(block,blockSize, m, processSize, n,count=1):
    flg=0
    if len(processSize)>1: 
        iterlist=list(subset_sum(processSize, 10))[0]
        for a in iterlist:
            #print (a)
            processSize.remove(a)  # remove the element from weights array once room is allotted.
    else :
        
        iterlist=processSize
        flg=1
    #print (numbers)
          
    n=len(iterlist)
    # Stores rooms id of the  
    # room allocated to a weight  
    allocation = [-1] * n  
    # Initially no room is assigned to any weights 
    # pick each weight and find suitable room  
    # according to its size ad assign to it  
    
    for i in range(n): 
        for j in range(m): 
            if blockSize[j] >= iterlist[i]: 
                  
                # allocate room j to p[i] weights  
                allocation[i] = j  
  
                # Reduce available weights in this room.  
                blockSize[j] -= iterlist[i]  
  
                break
    
    print (f"______________Iteration {count} _________________")
    print ('\n')
    print (f"Block Size {block}")
    print (f"Available Sizes {blockSize}")

    print(" sl No. weights carried      Block no.") 
    
    for i in range(n): 
        print(" ", i + 1, "         ", iterlist[i],  
                          "         ", end = " ")  
        if allocation[i] != -1:  
            print(block[allocation[i]])  
        else: 
            print("Not Allocated") 
           
        
    #print (processSize)
    if flg:
        for a in iterlist:
            #print (a)
            processSize.remove(a)
    
    if len(processSize)>0 :
        #recurrive function to perform iteration of filling the weights in array is exhausted
        return firstFit(block,blockSize, m, processSize, n ,count+1)  
        
```

```python
"""Best Fit Decreasing (BFD): First order the items by size, from largest to smallest, then run
the Best Fit Algorithm."""

def bestFit(bb_block,bb_blockSize, bb_m, bb_processSize, bb_n,bb_count=1): 
    bb_flg=0
    if len(bb_processSize)>1: 
        bb_iterlist=list(subset_sum(bb_processSize, 10))[0]
        for a in bb_iterlist:
            #print (a)
            bb_processSize.remove(a)   # remove the element from weights array once room is allotted.
    else :
        bb_iterlist=bb_processSize
        bb_flg=1
    #print (numbers)
        
    bb_n=len(bb_iterlist) 
    # Stores room id of the rooms  
    # allocated to a weight element  
    bb_allocation = [-1] * bb_n  
    # pick each room and find suitable  
    # weights according to its size and  
    # assign to it 
    
    for bb_i in range(bb_n): 
        # Find the best fit block for 
        # current process  
        bb_bestIdx = -1
        for bb_j in range(bb_m):
            #print (i,j,roomSize,weightSize)
            if bb_blockSize[bb_j] >= bb_iterlist[bb_i]: 
                #print (roomSize[j], weightSize[i] )
                if bb_bestIdx == -1:  
                    bb_bestIdx = bb_j  
                elif bb_blockSize[bb_bestIdx] > bb_blockSize[bb_j]:  
                    bb_bestIdx = bb_j 
       
        # If we could find a room for current weight  
        if bb_bestIdx != -1: 
            # allocate room j to p[i] weight  
            bb_allocation[bb_i] = bb_bestIdx  
  
            # Reduce available space in this room.  
            bb_blockSize[bb_bestIdx] -= bb_iterlist[bb_i] 
        
        
    print (f"______________Iteration {bb_count} _________________")
    print ('\n')
    print (f"Block Size {bb_block}")
    print (f"Available Sizes {bb_blockSize}")
    print(" sl No. weights carried      Block no.") 
    for bb_i in range(bb_n): 
        print(bb_i + 1, "         ", bb_iterlist[bb_i],  
                                end = "         ")  
        
        if bb_allocation[bb_i] != -1:  
            print(bb_block[bb_allocation[bb_i]])  
        else: 
            print("Not Allocated") 
            
    if bb_flg:
        for a in bb_iterlist:
            #print (a)
            bb_processSize.remove(a)
    if len(bb_processSize)>0 :
        #recurrive function to perform iteration of filling the weights in array is exhausted
        return  bestFit(bb_block,bb_blockSize, bb_m, bb_processSize, bb_n,bb_count+1)
```

```python
#weight send in the robot of 10 kgs as follows [9,1],[7,2,1] and so on
#function define below will be selecting the weights with sum equal to 10kgs

def subset_sum(numbers, target, partial=[], partial_sum=0):
    if partial_sum == target:
        yield partial
    if partial_sum >= target:
        return
    for i, n in enumerate(numbers):
        remaining = numbers[i + 1:]
        yield from subset_sum(remaining, target, partial + [n], partial_sum + n)
    
```
